# Histogram

```python
ages =[10,15,20,16,30,32,35,40,45,56]
plt.hist(ages,bins=5,color='b',edgecolor='r')#وزع الداتا  و بيشوف عدد الاشخاص بين لبعمر كذا و كذا 
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

```python
ages =[10,15,20,16,30,32,35,40,45,56]
bins=[10,20,30,40,50]
plt.hist(ages,bins=bins,color='b',edgecolor='r')
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

```python
import numpy as np

ages =[10,15,20,16,30,32,35,40,45,56]
plt.hist(ages,bins=5,color='b',edgecolor='r') 
mean=np.mean(ages)        
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

print(mean)
plt.axvline(mean,color='r',linewidth=5) # simulating mean use fun=axvline(....,color='')
```

```python
import matplotlib .pyplot as plt
plt.style.use('ggplot')
```

```python
data=[80,20,10,50]
labels=['pyton','js','HTML','c++']
plt.pie(data,labels=labels)
plt.title('A Pie Chart')
```

```python
data=[80,20,10,50]
labels=['pyton','js','HTML','c++']

colors=['yellow','red','green','blue']
plt.pie(data,labels=labels,colors=colors,wedgeprops={'edgecolor':'black'})
plt.title('A Pie Chart')
```

```python
data=[80,20,10,50]
labels=['pyton','js','HTML','c++']

explode=[0,0,0.2,0]

plt.pie(data,labels=labels,explode=explode,wedgeprops={'edgecolor':'black'})
plt.title('A Pie Chart')
```

```python
data=[80,20,10,50]
labels=['pyton','js','HTML','c++']

explode=[0,0,0.2,0]

plt.pie(data,labels=labels,explode=explode,shadow=True,wedgeprops={'edgecolor':'black'})
plt.title('A Pie Chart')
```

```python
data=[80,20,10,50]
labels=['pyton','js','HTML','c++']

explode=[0,0,0.2,0]

plt.pie(data,labels=labels,explode=explode,shadow=True,startangle=90,wedgeprops={'edgecolor':'black'})
plt.title('A Pie Chart')
```

```python
data=[80,20,10,50]
labels=['pyton','js','HTML','c++']

explode=[0,0,0.2,0]

plt.pie(data,labels=labels,explode=explode,shadow=True,autopct='%1.1f%%',startangle=90,wedgeprops={'edgecolor':'black'})
plt.title('A Pie Chart')
```

```python
data=[80,20,10,50]
labels=['pyton','js','HTML','c++']

explode=[0,0,0.2,0]

plt.pie(data,labels=labels,explode=explode,shadow=True,rotatelabels=True,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
plt.title('A Pie Chart')
```
